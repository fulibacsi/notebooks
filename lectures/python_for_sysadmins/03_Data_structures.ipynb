## Complex data structures

### List

Lists are ordered mutable storage solution for diverse data. Lists stores references for the items in the list.

```python
pi = 3.14
my_list = [1, 2, 'spam', ['eggs', 'ham'], 'spam', pi]
# lists are ordered, and you can access an item using []
my_list[1] = 4
print(my_list)
```

You can add or remove elements from a list too with `append` and `remove`.
- `append` adds an item to the end of the list
- `extend` appends an iterable to the list
- `remove` removes the first appereance of the item

```python
my_list.append('new item')
print(my_list)
my_list.remove('spam')
print(my_list)
my_list.extend(['baked beans', 'eggs', 'spam'])
print(my_list)
```

### Tuple
Tuples are ordered immutable list of values.  
Typically tuples are used to store data pairs, like (page, line) locations in a document.

```python
my_tuple = ('spam', 'eggs', 'bacon')
print(my_tuple)
```

Since tuples are immutable element setting is forbidden.

```python
my_tuple[1] = 4
```

### Dictionary
Dictionaries are - as their name suggests - key-value pairs.  
Typical use of dictionaries is storing information of people in an organization.

```python
my_dict = {'key1': 'spam', 2: 2, 'key3': ['s', 'p', 'a', 'm', 3]}
print(my_dict)
```

Keys must be immutable.

```python
my_dict2 = {[1, 2, 3]: 'a list'}
```

```python
person1 = {'name': 'Mario'}
person2 = {'name': 'Sonic'}
# so you can access information in a neat way:
print(person1['name'])
# we can add new key-value pairs like this:
person1['company'] = 'nintendo'
```
