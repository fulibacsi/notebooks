```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn
```

```python
def plot_function(f, df, name, ax=None):
    x = np.linspace(-10, 10, 2000)
    y = f(x)
    dy = df(x)
    
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.plot(x, y, 'b-', label=f'{name}(x)')
    ax.plot(x, dy, 'r--', label=f"{name}'(x)")
    
    ax.legend(fontsize="x-large")
    
    # set the x-spine
    ax.spines['left'].set_position('zero')

    # turn off the right spine/ticks
    ax.spines['right'].set_color('none')
    ax.yaxis.tick_left()

    # set the y-spine
    ax.spines['bottom'].set_position('zero')

    # turn off the top spine/ticks
    ax.spines['top'].set_color('none')
    ax.xaxis.tick_bottom()
    
    return ax
```

## Activation function

```python
def sigmoid(x):
    return 1. / (1. + np.exp(-x))

def sigmoid_prime(x):
    return sigmoid(x) * (1 - sigmoid(x))
    
plot_function(sigmoid, sigmoid_prime, 'sigmoid');
```

```python
def tanh(x):
    return np.tanh(x)

def tanh_prime(x):
    return 1. - np.tanh(x) ** 2

plot_function(tanh, tanh_prime, 'tanh');
```

```python
def linear(x):
    return x

def linear_prime(x):
    return np.ones(x.shape)

plot_function(linear, linear_prime, 'linear');
```

```python
def elu(x, alpha=0.01):
    y = x.copy()
    y[y < 0] = alpha * (np.exp(y[y < 0]) - 1.)
    return y

def elu_prime(x, alpha=0.01):
    y = np.ones(x.shape)
    y[x < 0] = elu(y[x < 0]) + alpha
    return y

# we cheat a bit with the parameters to make it more visible
plot_function(lambda x: elu(x, 1), lambda x: elu_prime(x, 0.1), 'ELU');
```

```python
def relu(x):
    y = x.copy()
    y[y < 0] = 0
    return y

def relu_prime(x):
    y = np.ones(x.shape)
    y[x < 0] = 0
    return y

plot_function(relu, relu_prime, 'ReLU');
```

```python
def leaky_relu(x, alpha=0.01):
    y = x.copy()
    y[y < 0] = alpha * y[y < 0]
    return y

def leaky_relu_prime(x, alpha=0.01):
    y = np.ones(x.shape)
    y[x < 0] = alpha
    return y

# we cheat a bit with the parameters to make it more visible
plot_function(lambda x: leaky_relu(x, 0.1), lambda x: leaky_relu_prime(x, 0.1), 'Leaky ReLU');
```

```python
def softmax(x, alpha=0.01):
    y = x.copy()
    y[y < 0] = alpha * y[y < 0]
    return y

def leaky_relu_prime(x, alpha=0.01):
    y = np.ones(x.shape)
    y[x < 0] = alpha
    return y

# we cheat a bit with the parameters to make it more visible
plot_function(lambda x: leaky_relu(x, 0.1), lambda x: leaky_relu_prime(x, 0.1), 'Leaky ReLU');
```
