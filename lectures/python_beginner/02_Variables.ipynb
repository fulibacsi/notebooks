## Variables

What if we want to store some values? We use the assignment (`=`) operator!

- __`a`__ is a number

```python
a = 3
print(a)
print(type(a))
```

- __`a`__ is a floating point number

```python
a = 3.14
print(a)
print(type(a))
```

- __`a`__ is a boolean value

```python
a = True # False
print(a)
print(type(a))
```

- __`a`__ is a character

```python
a = 'a'
print(a)
print(type(a))
```

- __`a`__ is a string

```python
a = 'spam'
print(a)
# or
a = "spam and eggs"
print(a)
# or
a = '''spam'
spam
sausage'''
print(a)
# or
a = """spam
spam
eggs
and spam"""
print(a)
print(type(a))
```

__According to strings:__  
There are some so called __`escape character`__s. They always starts with a __\\__ following by an another character, like:
- `\n`
- `\t`
- `\r`
- `\'`
- `\"`

Try them out!

```python
a = 'sp\nam'
print(a)
```
